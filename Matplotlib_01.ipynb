```python
# Librerias
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creacion de un Simple Line Plots
# Datos de Meses por cantidad de autos en trafico
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
trafico = [150,200,230,300,120,100]
# Creamos la linea de plot
plt.plot(meses,trafico)
# Mostrar
plt.show()
```

```python
# Creacion de un Simple Line Plots
# Datos de Meses por cantidad de autos en trafico
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
trafico = [150,200,230,300,120,100]
# Creamos la linea de plot
plt.plot(meses,trafico)
# Agregar los titulos
plt.xlabel('Meses') # Titulo en el eje de las x
plt.ylabel('Cantidad de Trafico') # Titulo en el eje de las y
plt.title ('Meses de Trafico') # Titulo de la imagen
# Mostrar
plt.show()
```

```python
# Creacion de un Simple Line Plots
# Datos de Meses por cantidad de autos en trafico
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
trafico = [150,200,230,300,120,100]
# Creamos la linea de plot
plt.plot(meses,trafico,marker = 'o', linestyle ='--', color = 'g')
# Agregar los titulos
plt.xlabel('Meses') # Titulo en el eje de las x
plt.ylabel('Cantidad de Trafico') # Titulo en el eje de las y
plt.title ('Meses de Trafico') # Titulo de la imagen
plt.grid(True)
# Mostrar
plt.show()
```

```python
# Creacion de un Simple Line Plots
# Datos de Meses por cantidad de autos en trafico
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
trafico_a = [150,200,230,300,120,100]
trafico_b = [10,30,20,90,60,70]
# Creamos la linea de plot
plt.plot(meses,trafico_a,marker = 'o', linestyle ='-', color = 'blue',label = 'Trafico A')
plt.plot(meses,trafico_b,marker = 'o',linestyle ='-',color = 'red',label ='Trafico B')
# Agregar los titulos
plt.xlabel('Meses') # Titulo en el eje de las x
plt.ylabel('Cantidad de Trafico') # Titulo en el eje de las y
plt.title ('Meses de Trafico Comparacion') # Titulo de la imagen
plt.legend() # muestra la leyenda
plt.grid(True) # muestra la grilla
# Mostrar
plt.show()
```

```python
# Creacion de un Simple Line Plots
# Datos de Meses por cantidad de autos en trafico
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
trafico_a = [150,200,230,300,120,100]
trafico_b = [10,30,20,90,60,70]
# Creamos la linea de plot
plt.plot(meses,trafico_a,marker = 'o', linestyle ='-', color = 'blue',label = 'Trafico A') # muestra punto
plt.plot(meses,trafico_b,marker = 's',linestyle ='--',color = 'red',label ='Trafico B') # muestra un cuadrado
# Agregar los titulos
plt.xlabel('Meses') # Titulo en el eje de las x
plt.ylabel('Cantidad de Trafico') # Titulo en el eje de las y
plt.title ('Meses de Trafico Comparacion') # Titulo de la imagen
plt.legend() # muestra la leyenda
plt.grid(True) # muestra la grilla
# Mostrar
plt.show()
```
